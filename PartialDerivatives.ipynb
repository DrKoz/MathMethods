# **5. Partial Derivatives**


Python can do both symbolic and numerical derivatives! Differentiation is done using the sympy library so we need to import that for this section along with the usual numpy and matplotlib.

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

Let's try a 1D example. Find the derivative of $f(x)=x^2-4x+3$ and evaluate it at x=1.




```python
# create a "symbol" called x
x = sym.symbols('x')
#write the function
f = x**2-4*x+3

#take the derivative with respect to x. 
df = f.diff(x)      

print('f(x)=',f)
print('df/dx=', df)

#Here is how to evaluate a function
print('f(1)=',f.subs(x,1))
print('f\'(1)=', df.subs(x,1))

#If you need floating point output (i.e.  decimal), use .evalf()
print('f(1)=',f.subs(x,1).evalf())
print('f\'(1)=', df.subs(x,1).evalf())


```

We can expand what we did previously for partial derivatives, taking a derivative of the function previously graphed. Let's evaluate at $f(1,-2)$.

```python
# create "symbols" called x and y
x , y = sym.symbols('x, y')
f = x ** 3 - 2*x*y**2
#set evaluation points
x0=1
y0=-2
 
#Differentiating partially w.r.t x and y
derivative_fx = f.diff(x)
derivative_fy = f.diff(y)
print('df/dx = ',derivative_fx)
print('df/dy = ',derivative_fy)


print('f(1,-2)=',f.subs([(x,x0),(y,y0)]))
print('df/dx=', derivative_fx.subs([(x,x0),(y,y0)]))
print('df/dy=', derivative_fy.subs([(x,x0),(y,y0)]))

#For second derivatives, include to variables in the diff() to indicate which partials
derivative_fxx = f.diff(x,x)      #d^2f/dx^2
derivative_fyy = f.diff(y,y)      #d^2f/dy^2
derivative_fxy = f.diff(x,y)      #d^2f/dxdy
print('d2f/dx2 = ',derivative_fxx)
print('d2f/dy2 = ',derivative_fyy)
print('d2f/dxdy = ',derivative_fxy)
```

It turns out that there's a way to convert the symbolic python to useful numeric entries. Converting many values to plot using .subs() would be really tedious. So the sympy library has a function called lambdify that converts the symbolic function to an evaluatable (hmmm...not sure this is a word) function.  

```python
# create "symbols" called x and y
x , y = sym.symbols('x, y')
f = x ** 3 - 2*x*y**2
#set evaluation points
x0=1
y0=-2
 
#Differentiating partially w.r.t x and y
derivative_fx = f.diff(x)
derivative_fy = f.diff(y)
print('df/dx = ',derivative_fx)
print('df/dy = ',derivative_fy)

f=sym.lambdify([x,y],f,'numpy')        # in lambdify, first entry is the variable(s), second is the symbolic function, numpy tells it to convert sympy functions to numpy functions in order to evaluate numerically

fx = sym.lambdify([x,y],derivative_fx,'numpy')
fy = sym.lambdify([x,y],derivative_fy,'numpy')
print('f(1,-2) =', f(x0,y0))
print('fx(1,-2) =',fx(x0,y0))
print('fy(1,-2) =', fy(x0,y0))
print('Ok, these match up!')
```

We can combine what we learned above and what we learned in linear algebra to build the Hessian matrix and determine maxima, minima, and saddle points.

First, all the derivatives are taken. Then, the 1st derivatives are set equal to zero and solved as a system of equations using sym.solve([eqns],[variables]). The equations must be entered such that they equal zero. Since we want the derivatives equal to zero, we can just put in the derivatives.

Then we have to loop through the critical points to determine max, min or saddle.

```python
# create "symbols" called x and y
x , y = sym.symbols('x, y')
f = x ** 2 +3*x+2- x*y**2
#set evaluation points
x0=1
y0=-2
 
#Differentiating partially w.r.t x and y
derivative_fx = f.diff(x)
derivative_fy = f.diff(y)
derivative_fxx = f.diff(x,x)      #d^2f/dx^2
derivative_fyy = f.diff(y,y)      #d^2f/dy^2
derivative_fxy = f.diff(x,y)      #d^2f/dxdy

fx = sym.lambdify([x,y],derivative_fx,'numpy')      #df/dx
fy = sym.lambdify([x,y],derivative_fy,'numpy')      #df/dy
fxx = sym.lambdify([x,y],derivative_fxx,'numpy')      #d^2f/dx^2
fyy = sym.lambdify([x,y],derivative_fyy,'numpy')      #d^2f/dy^2
fxy = sym.lambdify([x,y],derivative_fxy,'numpy')      #d^2f/dxdy

w=sym.solve([derivative_fx, derivative_fy],[x,y])    #find critical points. solve() assumes the functions (first derivatives, in this case) are equal to zero and it will solve for defined variables [x,y]

print(w, len(w)) #solutions and number of solutions
print(w[0][0])   # solutions are in the form w[i][j] where i counts the solution number and j stores the variables (for this x is j=0, y is j=1)

i=0
while i<len(w):
  a0=float(w[i][0])
  a1=float(w[i][1])
  H = np.array([[fxx(a0,a1), fxy(a0,a1)],[fxy(a0,a1), fyy(a0,a1)]])

  detH = np.linalg.det(H)
  print('H=\n',H)
  print('det(H)=', detH)
  if detH < 0:
    print(w[i], 'is a saddle point.')
  elif detH > 0:
    print(fxx(a0,a1), fyy(a0,a1))
    if(fxx(a0,a1) > 0 and fyy(a0,a1) > 0):
      print(w[i], 'is a minimum.')
    elif(fxx(a0,a1) > 0 and fyy(a0,a1) > 0):
      print(w[i], 'is a maximum.')
    else:
      print(w[i],'cannot be determined')
  else:
    print(w[i],'cannot be determined')
  print('  ')
  i=i+1

```
